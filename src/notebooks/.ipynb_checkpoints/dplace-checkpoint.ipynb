```python
import os
import pandas as pd
```

```python
# Define directory paths
parent_data_directory = "../../data/" # relative directory position to this file
binford_data_directory = os.path.join(parent_data_directory, 'dplace-data/datasets/Binford')

# mashed file path:
bin_path = os.path.join(binford_data_directory, 'merged_binford.csv')
```

```python
bindf = pd.read_csv(bin_path)
display(bindf)
```

```python
bvariables = bindf.loc[bindf['soc_id'] == 'B1', ['var_id', 'title', 'definition']]

bvariables_file = os.path.join(binford_data_directory, 'extr_variables.csv')
bvariables.to_csv(bvariables_file)
```
