### File selection
Brute force max file overlap years and period.
This is all but ideal, but I needed to ensure column naming convention compliance.

```python
def files_in_range(start_year:int, end_year:int, files_year:list) -> list:
    '''
    function takes a start and end year
    to return a list of files that contain year columns and data within the specified timeframe
    '''
    inrange_files = []

    for file in files_year:
        df = pd.read_csv(file, low_memory=False)
        marked = False

        for column in df.columns:
            if 'Year' in column:
                try:
                    col_min = pd.to_numeric(df[column], errors='coerce').min()
                    col_max = pd.to_numeric(df[column], errors='coerce').max()

                    if pd.notna(col_min) and pd.notna(col_max):
                        if 'Start' in column and col_min <= start_year:
                            marked = True
                        elif 'End' in column and marked and col_max >= end_year:
                            inrange_files.append(file)
                            break
                        elif not marked and col_min <= start_year and col_max >= end_year:
                            inrange_files.append(file)
                            break
                except (ValueError, TypeError):
                    continue

    return inrange_files
```

```python
def spread_comp(yrange:int, files_year:list) -> list:
    steps = (2025 - yrange - 1960) + 1
    file_spread = []

    for step in range(steps):
        start_year = 1960 + step
        end_year = 1960 + yrange + step
        
        inrange_files = files_in_range(start_year, end_year, files_year)

        file_spread.append([inrange_files, start_year, end_year])
        
    return file_spread
```

```python
# Will use between 8 and 20 years periods. I do not want to go less, so this will do.

iterno = 20-8 +1
overlap_maxes = []
maxes = []

print('Brute forcing max overlap...', '\n\n')
for i in range(iterno):
    print('Crunching ', 8+i, ' years range overlaps...')
    
    file_spread = spread_comp(8+i, files_year)
        
    # check max year overlap in current yrange
    max_overlap = 0
    max_o_index = 0
    for ind, spread in enumerate(file_spread):
        if len(spread[0]) > max_overlap:
            max_overlap = len(spread[0])
            max_o_index = ind

    cur_max_list = [8+i, max_overlap, file_spread[max_o_index]]
    print(f"Found {cur_max_list[1]} files as maximum overlap \n\n")
    maxes.append(cur_max_list)
```

#### Find best trusted
I manually select some datasets that look promising (homogenous community or place names) and then test the number of matches)

```python
from fuzzywuzzy import process

def find_best_trusted_fast(trusted_dfs, trusted_cols, other_df, other_col):
    '''
    Efficiently determines which trusted dataset has the highest overlap with others to minimize manual interventions.
    
    Returns the best trusted dataset based on name overlap.
    '''
    name_overlap_scores = {}

    for idx, trusted_df in enumerate(trusted_dfs):
        trusted_unique = set(trusted_df[trusted_cols[idx]].astype(str).str.strip().str.lower().unique())
        other_unique = set(other_df[other_col].astype(str).str.strip().str.lower().unique())

        overlap = len(trusted_unique.intersection(other_unique))
        missing = len(other_unique - trusted_unique)

        name_overlap_scores[idx] = {"overlap": overlap, "manual_interventions": missing}

    best_idx = min(name_overlap_scores, key=lambda x: name_overlap_scores[x]["manual_interventions"])
    best_trusted = trusted_dfs[best_idx]

    print("\nComparison of Trusted Datasets:")
    for idx, scores in name_overlap_scores.items():
        print(f"Dataset {idx+1}: {scores['overlap']} names match, {scores['manual_interventions']} manual interventions needed.")

    print(f"\nBest trusted dataset: {best_idx+1} (Only {name_overlap_scores[best_idx]['manual_interventions']} manual interventions needed)")
    
    return best_idx
```

#### Name Matching

```python
from fuzzywuzzy import process

def match_names(trusted_df, trusted_col, other_df, other_col):
    '''
    This function matches names between two dataframe columns, asking for confirmation.
    
    Returns a DataFrame with the original and final matched names for any subsequent checks.
    '''

    # Clean and strip names to keep mismatches to a minimum
    trusted_unique = set(trusted_df[trusted_col].astype(str).str.strip().str.lower().unique())
    other_unique = set(other_df[other_col].astype(str).str.strip().str.lower().unique())

    # Count manual matches needed
    manual_matches_needed = sum(1 for name in other_unique if name not in trusted_unique)
    print(f"⚠️ Manual intervention needed for {manual_matches_needed} names.\n")

    # Fuzzy match
    matched_other_to_trusted = {}

    for name in other_unique:
        if name in trusted_unique:
            matched_other_to_trusted[name] = name
        else:
            match = process.extractOne(name, list(trusted_unique))
            user_input = input(f"Match '{name}' to '{match[0]}'? (y/n/custom): ").strip().lower()

            if user_input == "y":
                matched_other_to_trusted[name] = match[0]
            elif user_input == "n":
                matched_other_to_trusted[name] = name
            else:
                matched_other_to_trusted[name] = user_input

    # Add unmatched to have a complete list
    for name in trusted_unique:
        if name not in matched_other_to_trusted.values():
            matched_other_to_trusted[name] = name

    matched_df = pd.DataFrame(list(matched_other_to_trusted.items()), columns=["Original", "Matched"])
    
    return matched_df
```
